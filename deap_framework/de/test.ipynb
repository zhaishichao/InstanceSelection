```python
import matplotlib.pyplot as plt
# 设置可显示中文宋体
plt.rcParams['font.family'] = 'STZhongsong'
x = ['[5,1]', '[20,2]', '[15,3]', '[25,4]', '[10,5]']
data = [5, 20, 15, 25, 10]
plt.title("基本柱状图")
#plt.grid(ls="--", alpha=0.5)
plt.bar(x, data)
for a, b in zip(x, data):
	plt.text(a, b + 0.1, '%.0f' % b, ha='center', va='bottom', fontsize=9)

```
