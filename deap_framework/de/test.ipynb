```python
import matplotlib.pyplot as plt
import numpy as np

# 设置图形大小
plt.figure(figsize=(15, 10))

# 生成随机数的数据
sample_size = 10000

# 1. 均匀分布
uniform_data = np.random.uniform(low=0.0, high=1.0, size=sample_size)
plt.subplot(2, 2, 1)
plt.hist(uniform_data, bins=30, color='skyblue', edgecolor='black')
plt.title("Uniform Distribution")

# 2. 正态分布
normal_data = np.random.normal(loc=0.0, scale=1.0, size=sample_size)
plt.subplot(2, 2, 2)
plt.hist(normal_data, bins=30, color='salmon', edgecolor='black')
plt.title("Normal Distribution")

# 3. 指数分布
exponential_data = np.random.exponential(scale=1.0, size=sample_size)
plt.subplot(2, 2, 3)
plt.hist(exponential_data, bins=30, color='lightgreen', edgecolor='black')
plt.title("Exponential Distribution")

# 4. 贝塔分布
beta_data = np.random.beta(a=2.0, b=5.0, size=sample_size)
plt.subplot(2, 2, 4)
plt.hist(beta_data, bins=30, color='plum', edgecolor='black')
plt.title("Beta Distribution")

# 调整布局和显示图形
plt.tight_layout()

plt.savefig("C://Users//zsc//Desktop//evolution computation//experiment//imgs"+"//"+"random_2",dpi=300, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import uniform, norm, expon, beta

# 设置图形大小
plt.figure(figsize=(15, 10))
sample_size = 1000

# 1. 均匀分布
uniform_data = np.random.uniform(0.0, 1.0, sample_size)
plt.subplot(2, 2, 1)
plt.hist(uniform_data, bins=30, density=True, color='skyblue', edgecolor='black', alpha=0.6)
x = np.linspace(0, 1, 100)
plt.plot(x, uniform.pdf(x, 0, 1), 'r-', lw=2, label="Envelope")
plt.title("Uniform Distribution")
plt.legend()

# 2. 正态分布
normal_data = np.random.normal(0.0, 1.0, sample_size)
plt.subplot(2, 2, 2)
plt.hist(normal_data, bins=30, density=True, color='salmon', edgecolor='black', alpha=0.6)
x = np.linspace(-4, 4, 100)
plt.plot(x, norm.pdf(x, 0, 1), 'r-', lw=2, label="Envelope")
plt.title("Normal Distribution")
plt.legend()

# 3. 指数分布
exponential_data = np.random.exponential(1.0, sample_size)
plt.subplot(2, 2, 3)
plt.hist(exponential_data, bins=30, density=True, color='lightgreen', edgecolor='black', alpha=0.6)
x = np.linspace(0, 10, 100)
plt.plot(x, expon.pdf(x, scale=1), 'r-', lw=2, label="Envelope")
plt.title("Exponential Distribution")
plt.legend()

# 4. 贝塔分布
beta_data = np.random.beta(2.0, 5.0, sample_size)
plt.subplot(2, 2, 4)
plt.hist(beta_data, bins=30, density=True, color='plum', edgecolor='black', alpha=0.6)
x = np.linspace(0, 1, 100)
plt.plot(x, beta.pdf(x, 2, 5), 'r-', lw=2, label="Envelope")
plt.title("Beta Distribution")
plt.legend()

# 调整布局和显示图形
plt.tight_layout()
plt.savefig("C://Users//zsc//Desktop//evolution computation//experiment//imgs"+"//"+"random_2",dpi=300, bbox_inches='tight')
plt.show()
```
