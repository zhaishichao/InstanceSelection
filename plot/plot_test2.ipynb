```python
import matplotlib.pyplot as plt

y3 = [
    0.95905250, 0.97148700, 0.98119650, 0.98514850, 0.99328800, 0.99343850, 
    0.99357600, 0.99366900, 0.99366900, 0.99370850, 0.99370850
]
# 创建图形
plt.figure(figsize=(10, 6))

# 绘制点线图
plt.plot(y3, marker='o', linestyle='-', color='b', markersize=5, linewidth=1)

# 添加标题和标签
plt.title('Y1 Values Point-Line Plot', fontsize=14)
plt.xlabel('Index', fontsize=12)
plt.ylabel('Y1 Value', fontsize=12)

# 调整x轴刻度
plt.xticks(range(0, len(y3), 5))

# 添加网格
plt.grid(True, linestyle='--', alpha=0.6)

# 显示图形
plt.tight_layout()
plt.show()
```
